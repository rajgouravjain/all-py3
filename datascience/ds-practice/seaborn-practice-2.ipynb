# Grid plots

```python
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(data= iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.stripplot)
```

```python
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
g = sns.FacetGrid(data=tips,row='sex',col='day')
g.map(sns.scatterplot,'total_bill','tip')
```

```python
g = sns.FacetGrid(data=tips,row='sex',col='day')
g.map(sns.kdeplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,row='sex',col='day')
g.map(sns.distplot,'total_bill',kde=False)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='day')
```

```python
g = sns.FacetGrid(data=tips,row='sex',col='day')
g.map(sns.scatterplot,'total_bill','tip')
```

# Regression plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill', y = 'tip', data=tips, col='day',hue='sex',aspect=1, height=5)
```
