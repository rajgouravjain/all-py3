```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("x vs x**2")
plt.subplot(1,2,2)
plt.plot(x,y**2)
plt.title("x vs x**4")
```

```python
fig = plt.figure()
axes1  = fig.add_axes([0.0,0.0,0.4, 0.8])
axes1.plot(x,y)
axes1.set_xlabel("X1")
axes1.set_ylabel("Y1")
axes1.set_title("SQUARE")
axes2  =  fig.add_axes([0.45,0.0,0.4, 0.8])
axes2.plot(y,x)
axes2.set_xlabel("X2")
axes2.set_ylabel("Y2")
axes2.set_title("SQRT")
```

```python
fig , axes =  plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("SQURE")
axes[0].set_ylabel("Y1")
axes[1].plot(y,x)
axes[1].set_title("SQRT")
axes[1].set_ylabel("Y2")
```
